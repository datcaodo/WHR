### 2. Exploring Relationships

```python
# Create a subset excluding the 'Country name' and 'year'
sub = df[['Life Ladder', 'Log GDP per capita', 'Social support', 'Healthy life expectancy at birth', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption', 'Positive affect', 'Negative affect']]
```

### Correlations:

```python
# Create a correlation matric using pandas 
sub.corr()
```

```python
# Create a correlation heatmap using matplotlib

plt.matshow(sub.corr())
plt.show()
```

```python
# Save figure 
plt.matshow(sub.corr())
plt.savefig("out.png")

# This will save the image in the working directory.
```

```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(9, 9)) # figure size 
plt.matshow(sub.corr(), fignum=f.number) # type of plot
plt.xticks(range(sub.shape[1]), sub.columns, fontsize=12, rotation=90) # x axis labels
plt.yticks(range(sub.shape[1]), sub.columns, fontsize=12) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=12) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

#### Create a correlation heatmap using seaborn:

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(sub.corr(), annot = True, ax = ax) # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.

#Saving the correlation heatmap
plt.savefig('correlation.png')  # Save as PNG
```

### Scatterplots:

```python
# Identifying the strongest correlations
sub.corr()
```

```python
# Create a scatterplot for 'Life Ladder' and 'Log GDP per capita' columns in seaborn

sns.lmplot(x = 'Life Ladder', y = 'Log GDP per capita', data = sub)
```

The correlation coefficient is 0.789 which is a sign that there is a strong positive correlation between the two columns. A strong correlation between Log GDP per capita and Life Ladder suggests that economic prosperity is closely linked to subjective well-being, highlighting the complex interplay between economic factors and quality of life outcomes on a societal level.

```python
sns.lmplot(x='Life Ladder', y='Log GDP per capita', data = sub)
plt.xlabel('Life Ladder')
plt.ylabel('Log GDP per capita')
plt.title('Scatterplot of positive correlation')
plt.savefig('scatterplot.png')  # Save the plot as a PNG file
plt.show()
```

### Categorial Plots:

##### Creating a categorical variable that splits the "Social support" column into categories

```python
# I am using a histogram to visualize the distribution of the variables. 
# This way, I can determine sensible categories for the Social support range. 
# I don't want to end up with too few observations in any one of the categories.
# The argument "kde" add a line that encompasses the distribution

sns.histplot(df['Social support'], bins = 20, kde = True)
```

#### Categorising the different ranges 

```python
df.loc[df['Social support'] < 0.6, 'Social support category'] = 'Low Support'
```

```python
df.loc[(df['Social support'] >= 0.6) & (df['Social support'] < 0.8), 'Social support category'] = 'Middle Support'
```

```python
df.loc[df['Social support'] >= 0.8, 'Social support category'] = 'High Support'
```

```python
df['Social support category'].value_counts(dropna = False)
```

```python
df.columns
```

```python
# Create a categorical plot in seaborn using the social support categories created above

sns.set(style="ticks")
cp = sns.catplot(x="Social support", y="Life Ladder", hue="Social support category", data=df)
```

```python
# fig = g.get_figure()
g.savefig("out.png") 

# Again, the image will be saved in the working directory. 
```

We know that the correlation coefficient is 0.72 between "Life Ladder" and "Social support". It signifies a robust and positive association between these two variables, suggesting that social support is an important contributor to overall life satisfaction or happiness levels.

### 3. Conclusion

#### Question: What are the key determinants of happiness according to the World Happiness Report 2024?

##### Answer: The key determinants of happiness according to the World Happiness Report 2024 are 'Log GDP per capita', 'Social support' and 'Healthy life expectancy at birth'.

#### Hypothesis for further testings: 

1. Economic Prosperity and Happiness: Countries with higher GDP per capita experience higher levels of happiness, as measured by the Life Ladder index.
2. Social Support and Happiness: Societies with stronger social support networks, indicated by metrics such as perceived social support or social capital, exhibit higher levels of happiness.
3. Health and Happiness: Longer healthy life expectancy is positively associated with higher levels of happiness, as individuals in countries with better health outcomes report greater life satisfaction.
